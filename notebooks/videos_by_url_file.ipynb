```python
import pandas as pd
import os
```

```python
def split_excel_to_txt_by_videotype(input_file="../table_tennis_data/metadata/game.xlsx", column="VideoType"):
    # Read Excel
    df = pd.read_excel(input_file, sheet_name="game")

    # Verify the column exists
    if column not in df.columns:
        raise ValueError(f"Column '{column}' not found in Excel file.")

    # Group rows by VideoType
    for video_type, group_data in df.groupby(column):

        video_type = str(video_type)

        # Output text file
        output_path = os.path.join(f"{video_type}.xlsx")

        dummy_group_data = group_data.iloc[:, [1, 2]]
        # Write to txt (tab-separated or space-separated)
        dummy_group_data.to_excel(output_path)

        print(f"Saved: {output_path}")

    print("\nCompleted generating TXT files!")
```

```python
split_excel_to_txt_by_videotype()
```
